##**ANSWER SIX**

Train split : This part of the dataset is used for training the ML model and here we note the accuracy of the model.

Validation split : Now, we will see if by changing some of the parameters in the model (hyper-parameter tuning) that the model is some increase in the accuracy.

Test split : This part tests whether the trained model can predict the output on the unseen dataset.

  Example: If we prepare for a football match there are three phases:

  1. Training phase (Train split) : Here, we train with our teammates and get guidance from our coaches.
  2. Friendly matches (Validation split) : Here, we play matches against other teams or in ourself by making teams to catch our mistakes before main matches.
  3. Main match (Test split) : Here, we see different teams from different areas or regions some with whom we haven;t played a single match. These types of matches tests our training and exercises that we have done earlier.
